```python

import csv
import numpy as np
import matplotlib.pyplot as plt

file = open("ex1data1.csv")
fileObject = csv.reader(file)


m=97
n=1
X = np.zeros((m,n))
Y = np.zeros((m,n))

index=0
for x,y in fileObject:
    X[index]=x
    Y[index]=y
    index=index+1

plt.scatter(X,Y)
plt.xlabel("Population of a city")
plt.ylabel("Profit of a food truck in that city")
```

```python
def computeCost(X,Y,theta):
    m = X.shape[0]
    y_predicted = np.dot(X,theta)
    #print(y_predicted.shape)
    squaredError = np.square(np.subtract(y_predicted,Y))
    meanSquaredError = np.sum(squaredError)/(2*m)
    return meanSquaredError   
```

```python
theta = np.zeros((n+1,1))
X=np.c_[np.ones(m),X]

print(computeCost(X,Y,theta))
alpha=0.01
plt.scatter(X[:,1],Y)
plt.plot(X[:,1],np.dot(X,theta),color='red')

for i in range(1500):
    error = np.dot(X,theta)- Y
    grad0= sum(error*np.reshape(X[:,0],(X[:,0].shape[0],1)))/m
    grad1= sum(error*np.reshape(X[:,1],(X[:,1].shape[0],1)))/m

    theta[0]=theta[0]-alpha*grad0
    theta[1]=theta[1]-alpha*grad1

print(theta)

plt.scatter(X[:,1],Y)
plt.plot(X[:,1],np.dot(X,theta),color='red')

print(computeCost(X,Y,theta))
 
```
